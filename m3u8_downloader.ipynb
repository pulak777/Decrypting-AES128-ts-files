```python
import m3u8
import requests
```

```python
file_name = "dynamic_programming"
url = "https://media.prog.io/file/DPSERIES/dynamic_programming.m3u8"
domain_url = "https://media.prog.io/file/DPSERIES/"
r = requests.get(url)
print(r.text)
```

```python
m3u8_master = m3u8.loads(r.text)
m3u8_master.data
```

```python
playlist_url = m3u8_master.data['playlists'][2]['uri']
playlist_url = domain_url + playlist_url
playlist_url
```

```python
r = requests.get(playlist_url)
playlist = m3u8.loads(r.text)
playlist.data
```

```python
i = 0
with open(str(file_name) + ".ts", 'wb') as f:
    for segment in playlist.data['segments']:
        print(i+1)
        url = segment['uri']
        r = requests.get(domain_url + url)
        f.write(r.content)
        i = i+1
```

```python
import json
with open(str(file_name) + ".txt", 'a') as f:
    f.write(json.dumps(playlist.data['segments'][0]['key']).replace(", ", ",\n"))
```
